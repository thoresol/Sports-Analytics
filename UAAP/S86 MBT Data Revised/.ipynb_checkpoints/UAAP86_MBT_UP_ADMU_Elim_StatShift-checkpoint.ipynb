```python
def home_away_elim_shift(home_r1_orig, away_r1_orig, home_r2_orig, away_r2_orig):
    home_r1 = home_r1_orig.copy()
    home_r2 = home_r2_orig.copy()
    away_r1 = away_r1_orig.copy()
    away_r2 = away_r2_orig.copy()
    for df in [home_r1, home_r2, away_r1, away_r2]:
        df[['fieldgoal_m','fieldgoal_a']] = df['fieldgoal_ma'].str.split('/', expand=True)
        df[['2pt_m','2pt_a']] = df['2pt_ma'].str.split('/', expand=True)
        df[['3pt_m','3pt_a']] = df['3pt_ma'].str.split('/', expand=True)
        df[['freethrow_m','freethrow_a']] = df['freethrow_ma'].str.split('/', expand=True)
        df['fieldgoal_m'] = pd.to_numeric(df['fieldgoal_m'])
        df['fieldgoal_a'] = pd.to_numeric(df['fieldgoal_a'])
        df['2pt_m'] = pd.to_numeric(df['2pt_m'])
        df['2pt_a'] = pd.to_numeric(df['2pt_a'])
        df['3pt_m'] = pd.to_numeric(df['3pt_m'])
        df['3pt_a'] = pd.to_numeric(df['3pt_a'])
        df['freethrow_m'] = pd.to_numeric(df['freethrow_m'])
        df['freethrow_a'] = pd.to_numeric(df['freethrow_a'])
    home_away_totals_r1 = pd.concat([home_r1[home_r1["jersey_no"]=="Totals"],away_r1[away_r1["jersey_no"]=="Totals"]], ignore_index=True)
    home_away_totals_r2 = pd.concat([home_r2[home_r2["jersey_no"]=="Totals"],away_r2[away_r2["jersey_no"]=="Totals"]], ignore_index=True)
    home_away_totals_r1.loc[2] = home_away_totals_r1.loc[0] + home_away_totals_r1.loc[1]
    home_away_totals_r2.loc[2] = home_away_totals_r2.loc[0] + home_away_totals_r2.loc[1]
    home_away_totals_r1 = home_away_totals_r1.drop(columns=["jersey_no","name","min","fieldgoal_ma","2pt_ma","3pt_ma","freethrow_ma","point_diff","fieldgoal_pct","2pt_pct","3pt_pct","freethrow_pct"])
    home_away_totals_r2 = home_away_totals_r2.drop(columns=["jersey_no","name","min","fieldgoal_ma","2pt_ma","3pt_ma","freethrow_ma","point_diff","fieldgoal_pct","2pt_pct","3pt_pct","freethrow_pct"])
    print(home_away_totals_r1,'\n',home_away_totals_r2)
    home_away_elim_shift = []
    home_away_totals_r1_colhead = list(home_away_totals_r1.columns)
    for i in home_away_totals_r1_colhead:
        home_away_elim_shift.append(float(
            100*(4/5)*(home_away_totals_r2[i][0] - home_away_totals_r2[i][1])/home_away_totals_r2[i][2]
             - 100*(home_away_totals_r1[i][0] - home_away_totals_r1[i][1])/home_away_totals_r1[i][2]
                                         ))
    print(home_away_elim_shift)
    home_away_elim_shift_df = pd.DataFrame({'stat':home_away_totals_r1_colhead,'pct_shift':home_away_elim_shift}).sort_values('pct_shift')
    print(home_away_elim_shift_df)
    plt.figure(figsize = (10, 10), facecolor='white')
    plt.barh(home_away_elim_shift_df['stat'], home_away_elim_shift_df['pct_shift'], color='maroon')
    return plt.show()
```

```python
home_away_elim_shift(up_r1, admu_r1, up_r2, admu_r2)
```
