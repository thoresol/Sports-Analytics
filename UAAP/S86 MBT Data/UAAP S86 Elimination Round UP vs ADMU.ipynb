```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
up_r1 = pd.read_csv("UAAP_86_MBT_R1_UP_ADMU_3.csv")
pd.set_option('display.max_columns', None)
up_r1.tail()
```

```python
admu_r1 = pd.read_csv("UAAP_86_MBT_R1_UP_ADMU_2.csv")
admu_r1.tail()
```

```python
up_r2 = pd.read_csv("UAAP_86_MBT_R2_UP_ADMU_2.csv")
up_r2.tail()
```

```python
admu_r2 = pd.read_csv("UAAP_86_MBT_R2_UP_ADMU_3.csv")
admu_r2.tail()
```

```python
up_admu_totals_r1 = pd.concat([up_r1[up_r1["jersey_no"]=="Totals"],admu_r1[admu_r1["jersey_no"]=="Totals"]], ignore_index=True)
up_admu_totals_r1.head()
```

```python
up_admu_totals_r2 = pd.concat([up_r2[up_r2["jersey_no"]=="Totals"],admu_r2[admu_r2["jersey_no"]=="Totals"]], ignore_index=True)
up_admu_totals_r2.head()
```

```python
up_admu_totals_r1.loc[2] = up_admu_totals_r1.loc[0] + up_admu_totals_r1.loc[1]
up_admu_totals_r1.head()
```

```python
up_admu_totals_r2.loc[2] = up_admu_totals_r2.loc[0] + up_admu_totals_r2.loc[1]
up_admu_totals_r2.head()
```

```python
up_admu_totals_r1 = up_admu_totals_r1.drop(columns=["pos_neg"])
```

```python
up_admu_totals_r2 = up_admu_totals_r2.drop(columns=["jersey_no","name","min","fieldgoal_ma","2pt_ma","3pt_ma","freethrow_ma","pos_neg"])
```

```python
up_admu_totals_r1_colhead = list(up_admu_totals_r1.columns)
```

```python
up_admu_totals_r2_colhead = list(up_admu_totals_r2.columns)
```

```python
up_admu_r1_swing = []
for i in up_admu_totals_r1_colhead:
    up_admu_r1_swing.append([i, 100*(up_admu_totals_r1[i][0] - up_admu_totals_r1[i][1])/up_admu_totals_r1[i][2]])
up_admu_r1_swing
```

```python
up_admu_r2_swing = []
for i in up_admu_totals_r2_colhead:
    up_admu_r2_swing.append([i, 100*(up_admu_totals_r2[i][0] - up_admu_totals_r2[i][1])/up_admu_totals_r2[i][2]])
up_admu_r2_swing
```

```python
up_admu_elim_shift = []
for i in up_admu_totals_r2_colhead:
    up_admu_elim_shift.append(100*(up_admu_totals_r2[i][0] - up_admu_totals_r2[i][1])/up_admu_totals_r2[i][2]
                              - 100*(up_admu_totals_r1[i][0] - up_admu_totals_r1[i][1])/up_admu_totals_r1[i][2])
up_admu_elim_shift
```

```python
up_admu_elim_shift_df = pd.DataFrame(np.array(up_admu_elim_shift),columns=['stat','pct_shift'])
up_admu_elim_shift_df.set_index("stat", inplace=True)
up_admu_elim_shift_df
```

```python
plt.figure(figsize = (10, 10), facecolor='white')
plt.barh(up_admu_totals_r1_colhead, up_admu_elim_shift, color='maroon')
plt.show()
```
